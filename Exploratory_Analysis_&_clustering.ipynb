```python
import numpy as np
from tabulate import tabulate
import pandas as pd
from sklearn.cluster import KMeans
import random
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#data cleaning and extracting relevant features
df1 = pd.read_csv("food_cleaned_data.csv")
df=df1[["cook","diet_current_coded","eating_out","sports","exercise","fav_cuisine_coded","on_off_campus","pay_meal_out","fav_food","fruit_day","income"]]
df.dropna(axis=0,inplace=True)
df.to_csv("food_choices.csv")
df
```

```python
#Plotting Boxplot for cleaned data
sns.boxplot(data=df, palette="Set1").tick_params(labelsize=8.7)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#K-Means clustering on cleaned data
k = 3
kmeans = KMeans(n_clusters = k, random_state=0).fit(df)
df['Cluster']=kmeans.labels_
```

```python
k = 3

# Create subplots
fig, axes = plt.subplots(1, k, sharey=True, figsize=(15, 5))  # Adjust figsize as needed
axes[0].set_ylabel('Coded Values', fontsize=20)

# Loop through each cluster to create individual boxplots
for i in range(k):
    plt.sca(axes[i])  # Set current axis
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
    # Create a boxplot for the ith cluster, dropping the 'Cluster' column
    sns.boxplot(palette="Set1", data=df[df['Cluster'] == i].drop('Cluster', axis=1), ax=axes[i])
    axes[i].tick_params(labelsize=8.7)  # Adjust the font size of the ticks

# Display the plot
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 60)
    res = kmeans.fit(df)
    print(res)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
